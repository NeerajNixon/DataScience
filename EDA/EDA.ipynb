# AMPHIBIANS DATASET

```python

with open('/content/drive/MyDrive/Colab Notebooks/data/dataset.csv','r') as fin:
    data_3 = fin.read().splitlines(True)
    j = 0
    for line in data_3:
       i = 0
       for char in line: 
          if char == ';':
            line = line[:i] + ',' + line[i+1:]
            
          i+=1

       data_3[j] = line
       j+=1

with open('/content/drive/MyDrive/Colab Notebooks/data/dataset_new.csv', 'w') as fout:
    fout.writelines(data_3[1:])
```
